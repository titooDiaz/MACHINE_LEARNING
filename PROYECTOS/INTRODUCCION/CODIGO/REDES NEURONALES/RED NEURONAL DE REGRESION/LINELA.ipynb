```python
import pandas as pd
datos = pd.read_csv('datos.csv')
```

```python
datos.head()
```

```python
x = datos['celsius'].values #deveulva un arreglo
y = datos['fahrenheit']. values #deveuelve un array
```

```python
import tensorflow as tf

modelo = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(1,)),# numero de entradas
    tf.keras.layers.Dense(units=1)#la salida tambien es de un solo dato
])

modelo.compile(
    optimizer = tf.keras.optimizers.Adam(.1), #para que el modelo sepa que procesor seguir
    loss = 'mean_squared_error'
)
```

```python
entrenamiento = modelo.fit(x, y, epochs=1000)#las epocas nos dicen cuantas vieltas da
```

```python
import matplotlib.pyplot as plt

plt.xlabel('nuemro de epoca')
plt.ylabel('magnitud de perdida')
plt.plot(entrenamiento.history['loss'])
```

```python
c = 1230
prediccion = modelo.predict([c])
print(f'predccion: {prediccion[0]}')
```
