```python
from sklearn.model_selection import train_test_split

x= datos.data
y= datos.target

x_ent, x_pru, y_ent, y_pru = train_test_split(x,y, test_size=.5)
```

```python
from sklearn import svm

modelo = svm.SVC()
modelo.fit(x_ent, y_ent)
```

```python
from sklearn import metrics
predicciones=modelo.predict(x_pru)
print("exactitud: ", metrics.accuracy_score(y_pru, predicciones))
print("precision: ", metrics.precision_score(y_pru, predicciones))
```

```python
#cual es el que mas nos conviene?

kernels = ["linear", "rbf", "sigmoid"]
gamma = [1, .01, .001, .0001, .00001]
for kernel in kernels:
    for gammas in gamma:
        modelo = svm.SVC(kernel=kernel)
        modelo.fit(x_ent, y_ent)
        predicciones=modelo.predict(x_pru)
        print(f" {kernel}, {gamma} exactitud: ", metrics.accuracy_score(y_pru, predicciones))
        print(f" {kernel}, {gamma} precision: ", metrics.precision_score(y_pru, predicciones)) 

#podemos observar que a la lineal no le afecta

# el que mejor unciona en este caso es el lineal
```

```python
modelo = svm.SVC(kernel='linear')
modelo.fit(x_ent, y_ent)
predicciones=modelo.predict(x_pru)
print(f" {kernel}, {gamma} exactitud: ", metrics.accuracy_score(y_pru, predicciones))
print(f" {kernel}, {gamma} precision: ", metrics.precision_score(y_pru, predicciones)) 
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_pru, predicciones))
```

```python
from sklearn.metrics import confusion_matrix
import pandas as pd
pd.DataFrame(confusion_matrix(y_pru,predicciones))
```
